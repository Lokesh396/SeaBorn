```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
```

```python
plt.plot()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.boxplot(x='size',data=tips)
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
plt.savefig('boxplot.pdf')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='husl')
plt.savefig('boxplot1.jpg')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,order=['Sun','Sat','Fri','Thur'])
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.boxplot(data=iris)
```

```python
sns.boxplot(data=iris,orient='horizontal',color='red',saturation=0.2)
plt.savefig('boxplot2.png')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,palette='winter_r')
plt.savefig('boxplot3.jpg')
```
