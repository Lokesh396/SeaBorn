```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
flights=flights.pivot('month','year','passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights,row_cluster=False)
plt.savefig('clustermap.pdf')
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=1)
plt.savefig('clustermap1.png')
```

```python
#in clustermap cmap is simlar to  palatte 
```

```python
sns.clustermap(flights,cmap='coolwarm',figsize=(10,9))
```

```python
#standardize across col or rows
```

```python
sns.clustermap(flights,standard_scale=1)
```

```python
#normalizing data set
```

```python
sns.clustermap(flights,z_score=0)
plt.savefig('clustermap2.jpg')
```
