```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal=np.random.rand(12,15)
```

```python
sns.heatmap(normal)
plt.savefig('heatmap.pdf')
```

```python
sns.heatmap(normal,annot=True,linewidth=3)
```

```python
sns.heatmap(normal,vmin=0,vmax=2,annot=True)
plt.savefig('heatmap1.jpg')
```

```python
sns.heatmap(normal,vmin=0,vmax=10,annot=True)
```

```python
flights=sns.load_dataset('flights')
flights.head(3)
```

```python
flights=flights.pivot('month','year','passengers')
```

```python
sns.heatmap(flights)
```

```python
?flights.pivot
```

```python
sns.heatmap(flights,annot=True,fmt='d',cmap='coolwarm')
plt.savefig('heatmap2.jpg')
```

```python
sns.heatmap(flights,annot=True,cmap='summer',fmt='d')
```

```python
sns.heatmap(flights,cmap='summer',annot=True,fmt='d',center=flights.loc['Jun',1954])
plt.savefig('heatmap3.jpg')
```
