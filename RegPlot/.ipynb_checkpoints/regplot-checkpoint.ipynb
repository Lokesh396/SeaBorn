```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.style.use('seaborn')
%matplotlib inline
```

```python
#reg plots are useful to see relationship between two variables
```

```python
tips=sns.load_dataset('tips')
tips.head(1)
```

```python
sns.regplot?
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,color='g')
plt.savefig('regplot.png')
```

```python
x=np.random.randn(115)
y=np.random.randn(115)
```

```python
sns.regplot(x=x,y=y)
plt.savefig('regplot1.png')
```

```python
sns.regplot(x=x,y=y,color='m',marker='d')
```

```python
sns.regplot(x=x,y=y,line_kws={'linewidth':2,'color':'g'})
plt.savefig('regplot2.png')
```

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.12,x_estimator=np.mean)
plt.savefig('regplot3.png')
```
