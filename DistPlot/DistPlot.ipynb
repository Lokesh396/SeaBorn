```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
plt.plot()
```

```python
num=np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='blue')
```

```python
label_dist=pd.Series(num,name='variable_x')
```

```python
sns.distplot(label_dist)
plt.savefig('displot.pdf')
```

```python
#warnings are about in the further releases distplot is going to be removed so it is better to use displot
```

```python
sns.distplot(label_dist,vertical=True)
plt.savefig('distplot1.png')
```

```python
sns.distplot(label_dist,hist=False)
```

```python
sns.distplot(label_dist,rug=True,hist=False)
```

```python
sns.distplot(label_dist,hist=True,rug=True,color='blue')
plt.savefig('distplot2.jpg')
```
