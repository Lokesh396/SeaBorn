```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#used for probalbility density functions
import matplotlib
matplotlib.style.use('ggplot')
```

```python
x=np.random.randn(10)
y=np.random.randn(10)
```

```python
sns.kdeplot(x,color='green',shade=True)
```

```python
sns.kdeplot(x,y,color='m')
```

```python
sns.kdeplot(y,x,shade=True,color='g')
```

```python
sns.kdeplot(x,y,cmap='summer',shade=True)
plt.savefig('kdeplot.png')
```

```python
sns.kdeplot(x,y,levels=15,cmap='summer',shade=True)
```

```python
iris=sns.load_dataset('iris')
iris.head(1)
```

```python
iris['species'].value_counts()
```

```python
setosa=iris.loc[iris.species=='setosa']
versicolor=iris.loc[iris.species=='versicolor']
```

```python
sns.kdeplot(setosa.petal_length,setosa.petal_width,cmap='summer',shade=True)
sns.kdeplot(versicolor.petal_length,versicolor.petal_width,cmap='husl')
plt.savefig('kdeplot2.png')
```

```python
sns.kdeplot(iris['petal_length'],iris['petal_width'],data=iris,hue='species',cmap='summer')
plt.savefig('kdeplot3.png')
```
