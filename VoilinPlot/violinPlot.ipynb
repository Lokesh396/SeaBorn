```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(1)
```

```python
sns.violinplot(x=tips['tip'])
```

```python
sns.violinplot(x=tips['total_bill'])
```

```python
sns.violinplot(x='size',y='total_bill',data=tips,hue='sex',palette='winter_r')
```

```python
sns.violinplot(x='size',y='total_bill',data=tips,hue='sex',palette='coolwarm',split=True)
plt.savefig('violinplot.pdf')
```

```python
sns.violinplot(x='tip',y='day',hue='smoker',palette='winter_r',data=tips,split=True)
plt.savefig('violinplot1.png')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',inner='quartile')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',inner='quartile',split=True,scale='count')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',inner='quartile',scale='count')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',inner='stick',scale='count')
plt.savefig('violinplot2.png')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='winter_r',scale='count',inner='stick',split=True,bw=1)
plt.savefig('violinplot3.png')
```
