```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import matplotlib
matplotlib.style.use('seaborn')
```

```python
plt.plot()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.stripplot(x='tip',data=tips)
```

```python
sns.stripplot(x=tips['total_bill'])
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='Dark2')
plt.savefig('stripplot.pdf')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
#jitter attribute is used to control the dispersioin of the plot
```

```python
sns.stripplot(x='sex',y='total_bill',hue='smoker',data=tips,jitter=0.07,palette='winter_r',linewidth=0.75)
plt.savefig('stripplot.jpg')
```

```python
sns.stripplot(x='sex',y='total_bill',hue='smoker',data=tips,jitter=0.07,palette='winter_r',linewidth=0.75,dodge=True)
```

```python
sns.stripplot(x='sex',y='total_bill',hue='smoker',data=tips,jitter=0.07,palette='winter_r',marker='D',linewidth=0.75,dodge=True,size=10,edgecolor='k')
```

```python
sns.stripplot(x='day',y='tip',data=tips)
sns.boxplot(x='day',y='tip',
            data=tips,palette='winter_r')
plt.savefig('boxstripplot.png')
```
