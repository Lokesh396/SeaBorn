```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head(1)
```

```python
sns.pairplot(iris,hue='species')
plt.savefig('paitplot.png')
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',palette='coolwarm')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
plt.savefig('pairgrid.png')
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species')
x=x.map(plt.scatter,color='k')
x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',x_vars=['petal_width','petal_length'],y_vars=['sepal_width','sepal_length'],palette='rainbow_r')
x=x.map(plt.scatter)
x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_lower(plt.scatter)
x=x.map_upper(sns.kdeplot)
x.add_legend()
plt.savefig('pairgrid2.png')
```

```python
x=sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter)
x.add_legend()
plt.savefig('pairgrid3.jpg')
```
