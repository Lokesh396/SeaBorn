```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.style.use('seaborn')
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
iris.head(1)
```

```python
sns.pairplot?
```

```python
x=sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='summer')
plt.savefig('pairplot.png')
```

```python
sns.pairplot(iris,vars=['petal_length','petal_width'],palette='summer',hue='species')
plt.savefig('pairplot1.png')
```

```python
sns.pairplot(iris,vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width']
             ,palette='summer',hue='species')
plt.savefig('pairplot2.png')
```

```python
sns.pairplot(iris,diag_kind='kde',palette='husl',hue='species',kind='reg')
plt.savefig('pairplot3.png')
```
