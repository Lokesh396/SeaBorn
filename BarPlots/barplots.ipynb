```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.columns
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='smoker',palette='winter_r')
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='smoker',palette='winter_r')
```

```python
sns.barplot(x='total_bill',y='day',data=tips,palette='spring')
```

```python
sns.barplot(x='day',y='tip',data=tips,palette='spring',order=['Sat','Fri','Sun','Fri'])
```

```python
sns.barplot(x='day',y='tip',data=tips,palette='spring',estimator=np.median)
```

```python
sns.barplot(x='smoker',y='tip',data=tips,estimator=np.median,hue='sex',palette='coolwarm')
plt.savefig('barplot1.jpg')
```

```python
sns.barplot(x='smoker',y='tip',data=tips,ci=99)
```

```python
sns.barplot(x='smoker',y='tip',data=tips,ci=30,palette='winter_r')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,palette='spring',capsize=0.6)
```

```python
sns.barplot(x='smoker',y='tip',data=tips,palette='husl',hue='sex')
```

```python
sns.barplot(x='size',y='tip',data=tips,palette='autumn',capsize=0.2,ci=12)
```

```python
sns.barplot(x='smoker',y='tip',data=tips,color='red')
```

```python
sns.barplot(x='smoker',y='tip',data=tips,color='red',saturation=0.4)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='smoker',palette='husl',ci=15,capsize=0.3,estimator=np.median,saturation=0.3)
plt.savefig('barplot2.jpg')
plt.savefig('barplot.pdf')
```
